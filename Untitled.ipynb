```python
import numpy as np
import matplotlib.pyplot as plt

# Time (days)
t = np.linspace(0, 7, 200)

# Different memory strengths
S_values = [0.5, 1, 2]
colors = ['blue', 'orange', 'green']

plt.figure(figsize=(8, 5))
for S, color in zip(S_values, colors):
    R = np.exp(-t / S) * 100
    plt.plot(t, R, label=f'S = {S}', linewidth=2, color=color)

# Labels and title
plt.title("Ebbinghaus's Forgetting Curve", fontsize=14)
plt.xlabel("Elapsed Time (days)", fontsize=12)
plt.ylabel("Memory Retention (%)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title="Memory Strength (S)")
plt.tight_layout()

# Save for Overleaf
plt.savefig("forgetting_curve_clean.png", dpi=300)
plt.show()
```
